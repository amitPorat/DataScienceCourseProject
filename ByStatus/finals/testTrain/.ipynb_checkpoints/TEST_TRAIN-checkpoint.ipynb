```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.animation as animation
pd.set_option('display.max_columns', None)
```

## Data Preparation

```python
Train = pd.read_excel('TRAIN.xlsx')
Test = pd.read_excel('TEST.xlsx')
```

```python
Small_Train=Train[Train.deleteRows==1].copy()
Small_Test=Test[Test.deleteRows==1].copy()
```

```python
del Train['deleteRows']
del Test['deleteRows']
del Small_Train['deleteRows']
del Small_Test['deleteRows']
```

```python
Train.dropna(inplace=True)
Test.dropna(inplace=True)
Train[Train.index %20 == 0]
Test[Test.index %20 == 0]
Train.reset_index(inplace=True,drop=True)
Test.reset_index(inplace=True,drop=True)

Small_Train.dropna(inplace=True)
Small_Test.dropna(inplace=True)
Small_Train[Small_Train.index %20 == 0]
Small_Test[Small_Test.index %20 == 0]
Small_Train.reset_index(inplace=True,drop=True)
Small_Test.reset_index(inplace=True,drop=True)

```

## Data plotting

```python
def visualizer(x, y, plot_type, title, xlabel, ylabel,color, rotation=False, rotation_value=60, figsize=(4,4)):
    plt.figure(figsize=figsize)
    
    
    if plot_type == "lineplot":
        sns.lineplot(data=x, label=xlabel,color=color)
    elif plot_type == "count":  
        sns.countplot(x)
   
    plt.title(title, fontsize=20)
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=18)
    plt.yticks(fontsize=13)
    if rotation == True:
        plt.xticks(fontsize=13,rotation=rotation_value)
    plt.show()
```

## Prediction

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import scikitplot as skplt
x_train = Train.drop(['name','status','Time',' Position Y',' Position Y.1',' Velocity Y',' Velocity Y.1'], axis = 1)
y_train = Train['status']
x_test = Test.drop(['name','status','Time',' Position Y',' Position Y.1',' Velocity Y',' Velocity Y.1'], axis = 1)
y_test = Test['status']

#x_train = x_train.abs()
#x_test = x_test.abs()


#x_train['absolute pos x'] = abs(x_train[' Position X'])
#x_train['absolute ver x'] = abs(x_train[' Velocity X'])
#x_train['absolute wrist x'] = abs(x_train[' Wrist Pos X'])
#x_train['absolute elbow x'] = abs(x_train[' Elbow pos X'])
#x_train['absolute pos x.1'] = abs(x_train[' Position X.1'])
#x_train['absolute ver x.1'] = abs(x_train[' Velocity X.1'])
#x_train['absolute wrist x.1'] = abs(x_train[' Wrist Pos X.1'])
#x_train['absolute elbow x.1'] = abs(x_train[' Elbow pos X.1'])

#x_test['absolute pos x'] = abs(x_test[' Position X'])
#x_test['absolute ver x'] = abs(x_test[' Velocity X'])
#x_test['absolute wrist x'] = abs(x_test[' Wrist Pos X'])
#x_test['absolute elbow x'] = abs(x_test[' Elbow pos X'])
#x_test['absolute pos x.1'] = abs(x_test[' Position X.1'])
#x_test['absolute ver x.1'] = abs(x_test[' Velocity X.1'])
#x_test['absolute wrist x.1'] = abs(x_test[' Wrist Pos X.1'])
#x_test['absolute elbow x.1'] = abs(x_test[' Elbow pos X.1'])

#x_train.drop([' Position X',' Velocity X',' Wrist Pos X',' Elbow pos X',' Position X.1',' Velocity X.1',' Wrist Pos X.1',' Elbow pos X.1'], axis = 1)
#x_test.drop([' Position X',' Velocity X',' Wrist Pos X',' Elbow pos X',' Position X.1',' Velocity X.1',' Wrist Pos X.1',' Elbow pos X.1'], axis = 1)
```

#dumb classifer

```python

from sklearn.dummy import DummyClassifier
dummy_classifier = DummyClassifier(strategy="uniform")
dummy_classifier.fit(x_train, y_train)
y_pred = dummy_classifier.predict(x_test)
print(classification_report(y_test, y_pred))
confusion_matrix(y_test, y_pred)
```

#knn

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
#print(knn.score(x_test,y_test))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```

#logisic reg

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```

#naive bayas

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
y_pred = gnb.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```

#decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier()
dectree.fit(x_train,y_train)
y_pred = dectree.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```

#ada_boost

```python
from sklearn.ensemble import AdaBoostClassifier
ab = AdaBoostClassifier()
ab.fit(x_train, y_train)
y_pred = ab.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```

#random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```

```python
from sklearn.neural_network import MLPClassifier
NNW = MLPClassifier()
NNW.fit(x_train,y_train)
y_pred = NNW.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
skplt.metrics.plot_confusion_matrix(
    y_test, 
    y_pred,
    figsize=(7,7))
```
